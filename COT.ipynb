```python
from groq import Groq
client = Groq(api_key="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")
cot_prompt = """
Solve this problem step by step
Problem: A store sells apples for $2 each. If you buy 5 or more, you get 20% off the total. How much would 7 apples cost?
Think through this step by step:
"""
response = client.chat.completions.create(
    model = "llama-3.1-8b-instant",
    messages = [
    {
        "role": "system",
        "content": "You are a helpful assistant that reasons step-by-step. "
                   "Provide your reasoning inside <thought> tags and "
                   "your final answer inside <answer> tags."
    },
    {"role": "user", "content": "How much would 7 apples cost?"}
],
    temperature = 0.7,
    max_tokens = 1024
)
print(response.choices[0].message.content)
```
